Importe a tabela dos 100 maiores municípios em relação ao PIB e responda as seguintes perguntas:

Quantos municípios estão no estado de São Paulo?
Qual a participação acumulada desses municípios?
Dicas e informações:

Fonte dos dados: https://www.ibge.gov.br/estatisticas/

```python
import pandas as pd
```

```python
file_path = 'assets_dados/PIB_100_maiores_cidades_2017.txt'
ibge_df = pd.read_csv(file_path, encoding="latin-1")
display(ibge_df)
```

```python
contador_estados = ibge_df['Estado'].value_counts()
contador_estados
```

```python
is_SP = ibge_df['Estado'] == 'SP'
is_SP #* Retorna uma series 

df_SP = ibge_df[is_SP] 
df_SP.head()
```

```python
num_participacoes = df_SP['Participação acumulada (%)'].sum()
num_participacoes
```
