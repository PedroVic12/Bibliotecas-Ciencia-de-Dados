# ***Tratando dados de itinerários***

Dados horários de intinerários, pede-se:

- Encontre e filtre as razões sociais que se repetem duas vezes;

- Encontre quais delas possuem intervalos em menos de uma hora.

https://dados.antt.gov.br/dataset/gerenciamento-de-autorizacoes

Uso do método .isin e da função pd.to_datetime

```python
import pandas as pd
```

```python
df = pd.read_csv('assets_dados/horarios.csv',sep = ';', engine='python', encoding="latin-1")
df
```

```python
df1 = df[['razao_social','horario']]
df1
```

### Encontrando as que aparecem 2 vezes

```python
filtro1 = df1['razao_social'].value_counts() == 2
filtro1
```

### Pegando apenas os valores "True"

```python
filtro1[filtro1]
```

### Listas das razões sociais que aparecem 2 vezes

```python
lista_filtrada = list(filtro1[filtro1].index)
lista_filtrada
```

```python
len(lista_filtrada)
```

```python
df1['razao_social'].isin(lista_filtrada)
```

```python
tabela_filtrada = df1[df1['razao_social'].isin(lista_filtrada)]
tabela_filtrada 
```

### Encontrando em quais horarios são menos de 1h

```python
df1['time'] = pd.to_datetime(df['horario'])
df1
```

```python
print('Media dos Horarios')
pd.pivot_table(tabela_filtrada, index='razao_social',values='time')
```

```python
serie_max = pd.pivot_table(tabela_filtrada, index='razao_social',values='time', aggfunc='max')
serie_max
```

```python
serie_min = pd.pivot_table(tabela_filtrada, index='razao_social',values='time', aggfunc='min')
serie_min
```

```python
serie_max - serie_min
```
