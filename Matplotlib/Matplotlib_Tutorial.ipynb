## Pyplot e Elementos Gráficos 

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
xa_array = np.linspace(0,2*np.pi,500)
ya_array = np.sin(xa_array)

xb_array = np.linspace(0,2*np.pi,500)
yb_array = np.sin(xb_array + np.pi/2)

#!Plotando o gráfico
plt.plot(xa_array,ya_array,'y-')
plt.plot(xb_array,yb_array,'r-')

#*Eixos
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

#Linhas de grade
plt.grid() 
```

```python
fig, ax = plt.subplots()

ax.plot(xa_array,ya_array,label = 'Função A')
ax.plot(xb_array,yb_array,label = 'Função B')
ax.set(title = 'Função Seno', xlabel = 'Eixo X',ylabel = 'Eixo Y')

plt.legend()
```

## Subplots

```python
xa_array = np.linspace(0,5,500)
ya_array = np.exp(xa_array)

fig, (ax1,ax2) = plt.subplots(2,2)
ax1[1].plot(xa_array,ya_array)
ax1[0].plot(xa_array+100,ya_array)
```

## Grafico de Barras

```python
valores = [20,50,10,30,70]
nomes = ['A','B','C','D','E']
valores2 = [10,20,30,20,45]

plt.bar(nomes,valores, color = 'r')
plt.bar(nomes,valores2, color = 'b')
```

```python
fig, ax = plt.subplots()
x = np.arange(5)

largura = 0.4
rect1 = ax.bar(x+largura/2,valores, width=largura, color = 'red', label = 'Gráfico 1')
rect1 = ax.bar(x-largura/2,valores2, width=largura, color = 'blue', label = 'Gráfico 1')
plt.legend()

ax.set_xticks(x)
ax.set_xticklabels(nomes)
```

## Gráfico de Pizza

```python
plt.figure(figsize=(8,8))
explodir = (0,0,0.2,0,0)

plt.pie(valores,labels=nomes, autopct='%%1.1.f%%', explode=explodir,shadow=True)

plt.show()
```

## Gráfico de Superfície

```python
A = np.arange(3)
malha = np.meshgrid(A,A)
z = np.array([[1,0,0],
              [0,1,0],
              [0,0,1]])
malha
```

```python
fig = plt.figure(figsize=(6, 6))

ax = fig.gca(projection = '3d')
surf = ax.plot_surface(malha[0],malha[1],z)
```

```python
pi = np.linspace(0,np.pi,50)
malha_2 = np.meshgrid(pi,pi)
z_2 = np.sin(malha_2[0])*np.sin(malha_2[1])
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection = '3d')
surf = ax.plot_surface(malha_2[0],malha_2[1],z_2, cmap = cm.coolwarm)

```
