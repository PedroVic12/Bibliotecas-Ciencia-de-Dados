Dada a equaÃ§Ã£o da circunferÃªncia, faÃ§a o grÃ¡fico.

$$ (x-a)^2 + (y-b)^2 = r^2  $$

r = 100

a = 100

b = 100

```python
"""
*Isolando o Y:
! (y-b)^2 = r^2 - (x-a)^2
! y = sqrt(r**2 - (x-a)**2) + b 
"""

from math import sqrt
r,a,b = 100,100,100

x = []
y = []

x = [i for i in range(2*r + 1)] #* 2*r + 1 = 201 â DIAMETRO!

y_x = lambda x: sqrt((r**2)- (x-a)**2) + b

y_x_negativo = lambda x: -sqrt((r**2)- (x-a)**2) + b

y.extend(list(map(y_x,x)))
y.extend(list(map(y_x_negativo,x)))

x.extend(x)

plt.plot(x,y)
```
